**Problem - 4: Matrix Operations**

**For the following arrays:**

A = np.array([[3, 4], [7, 8]])

B = np.array([[5, 3], [2, 1]])

1. Prove A * A^-1 = I:

```python
A = np.array([[3, 4], [7, 8]])

B = np.array([[5, 3], [2, 1]])

A_inv = np.linalg.inv(A)
identity_matrix = np.dot(A, A_inv)
print(identity_matrix)
```

2. Prove AB ≠ BA:

```python
AB = np.dot(A, B)
BA = np.dot(B, A)
print(f"AB: \n{AB}")
print(f"BA: \n{BA}")
```

3. Prove (AB)^T = B^T * A^T:

```python
transposed = np.transpose(np.dot(A, B))
transposed_check = np.dot(np.transpose(B), np.transpose(A))
print(f"(AB)^T: \n{transposed}")
print(f"B^T * A^T: \n{transposed_check}")
```

4. Solve the system of linear equations using inverse methods:

```python
A_eq = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_eq = np.array([-1, -3, 9])
X = np.linalg.inv(A_eq).dot(B_eq)
print(f"Solution to system: {X}")
```

```python
X_linalg = np.linalg.solve(A_eq, B_eq)  # Use np.linalg.solve() to solve the system

print("Solution using np.linalg.solve:", X_linalg)
```

**10.2 Experiment: How Fast is Numpy?**

**Element-wise Addition**

1. Using Python Lists:

```python
import time

# Create two lists of size 1,000,000
list1 = [i for i in range(1000000)]
list2 = [i for i in range(1000000)]

# Measure time for element-wise addition using a loop
start_time = time.time()
result_list = [list1[i] + list2[i] for i in range(len(list1))]
end_time = time.time()

print(f"Time taken for Python list element-wise addition: {end_time - start_time:.5f} seconds")
```

2. Using NumPy Arrays:

```python
import numpy as np

# Convert the lists into numpy arrays
array1 = np.array(list1)
array2 = np.array(list2)

# Measure time for element-wise addition using numpy
start_time = time.time()
result_numpy = array1 + array2
end_time = time.time()

print(f"Time taken for NumPy array element-wise addition: {end_time - start_time:.5f} seconds")
```

**Element-wise Multiplication**

1. Using Python Lists:

```python
# Measure time for element-wise multiplication using a loop
start_time = time.time()
result_list_multiplication = [list1[i] * list2[i] for i in range(len(list1))]
end_time = time.time()

print(f"Time taken for Python list element-wise multiplication: {end_time - start_time:.5f} seconds")
```

2. Using NumPy Arrays:

```python
# Measure time for element-wise multiplication using numpy
start_time = time.time()
result_numpy_multiplication = array1 * array2
end_time = time.time()

print(f"Time taken for NumPy array element-wise multiplication: {end_time - start_time:.5f} seconds")
```

**Dot Product**

1. Using Python Lists:

```python
# Measure time for dot product using a loop
start_time = time.time()
dot_product_list = sum(list1[i] * list2[i] for i in range(len(list1)))
end_time = time.time()

print(f"Time taken for Python list dot product: {end_time - start_time:.5f} seconds")
```

2. Using NumPy Arrays:

```python
# Measure time for dot product using numpy
start_time = time.time()
dot_product_numpy = np.dot(array1, array2)
end_time = time.time()

print(f"Time taken for NumPy array dot product: {end_time - start_time:.5f} seconds")
```

**Matrix Multiplication**

1. Using Python Lists:

```python
# Create two 1000x1000 matrices as lists of lists
matrix1 = [[i for i in range(1000)] for j in range(1000)]
matrix2 = [[i for i in range(1000)] for j in range(1000)]

# Measure time for matrix multiplication using nested loops
start_time = time.time()
result_matrix = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(1000)) for j in range(1000)] for i in range(1000)]
end_time = time.time()

print(f"Time taken for Python list matrix multiplication: {end_time - start_time:.5f} seconds")
```

2. Using NumPy Arrays:

```python
# Create two 1000x1000 matrices as numpy arrays
matrix1_np = np.random.rand(1000, 1000)
matrix2_np = np.random.rand(1000, 1000)

# Measure time for matrix multiplication using numpy
start_time = time.time()
result_matrix_numpy = np.dot(matrix1_np, matrix2_np)
end_time = time.time()

print(f"Time taken for NumPy array matrix multiplication: {end_time - start_time:.5f} seconds")
```
